```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path1 = 'train.csv'
path2 = 'valid.csv'
path3 = 'test.csv'

train = pd.read_csv(path1)
valid = pd.read_csv(path2)
test = pd.read_csv(path3)

print('train 데이터셋')
display(train.head())
print('valid 데이터셋')
display(valid.head())
print('test 데이터셋')
display(test.head())
```

```python
train['category'].value_counts()
```

```python
# headline_id와 body_id가 일치하는지 확인
train['same_id'] = train['headline_id'] == train['body_id']
train.loc[train['same_id']==False].head()
```

```python
# 날짜 변수 추가
train['headline_date'] = pd.to_datetime(train['headline_id'].str[3:11], format = '%Y%m%d')
train['body_date'] = pd.to_datetime(train['body_id'].str[3:11], format = '%Y%m%d')
train['date'] = np.where(train['headline_date']==train['body_date'],
                         train['headline_date'],
                         train[['headline_date','body_date']].min(axis=1)
                         )
train = train.drop(['headline_date','body_date'], axis=1)

# 요일 변수 추가
train['day'] = train['date'].dt.day_name()

#확인
display(train.head())
```
