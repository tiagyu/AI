````python
from common import request_chat_completion, stream_response_console
import requests
from bs4 import BeautifulSoup
````

````python
prompt = "안녕하세요"
response = request_chat_completion(
    prompt = prompt,
    stream = True
)
message = stream_response_console(response)
````

````python
def crawl_naver_entertainment(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "lxml")
    article = soup.find("div", class_= "newsct_article _article_body").text.strip()
    return article
````

````python
example_url = "https://n.news.naver.com/mnews/article/374/0000429845"
````

````python
crawl_naver_entertainment(example_url)
````
