```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://n.news.naver.com/mnews/article/016/0002442548"
```

```python
response = requests.get(url)
```

```python
response.text
```

```python
soup = BeautifulSoup(response.text, "lxml")
```

```python
article = soup.find("div", class_= "newsct_article _article_body").text.strip()
```

```python
print(article)
```

```python
def crawl_naver_entertainment(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "lxml")
    article = soup.find("div", class_= "newsct_article _article_body").text.strip()
    return article
```

```python
crawl_naver_entertainment(url)
```
